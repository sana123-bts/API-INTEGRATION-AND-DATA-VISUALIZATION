```python
import requests

api_url = "https://catfact.ninja/facts"
try:
    response = requests.get(api_url)
    response.raise_for_status()  # Raise an exception for bad status codes
    data = response.json()
    print("Successfully fetched data from Cat Facts API.")
    print(f"Number of facts received: {len(data['data'])}")
except requests.exceptions.RequestException as e:
    print(f"Error fetching data from Cat Facts API: {e}")
```

```python
import requests

api_url = "https://catfact.ninja/facts"
try:
    response = requests.get(api_url)
    response.raise_for_status()  # Raise an exception for bad status codes
    data = response.json()
    print("Successfully fetched data from Cat Facts API.")
    print(f"Number of facts received: {len(data['data'])}")
except requests.exceptions.RequestException as e:
    print(f"Error fetching data from Cat Facts API: {e}")
```

```python
import pandas as pd

# Convert the 'data' list into a pandas DataFrame
df = pd.DataFrame(data['data'])

# Inspect the DataFrame's structure and data types
display(df.info())

# Check for missing values
display(df.isnull().sum())

# Extract the length of each fact and store it in a new column
df['fact_length'] = df['fact'].apply(len)

# Display the first few rows with the new column
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram of the 'fact_length' column
sns.histplot(data=df, x='fact_length')
plt.xlabel("Fact Length")
plt.ylabel("Frequency")
plt.title("Distribution of Cat Fact Lengths")
plt.show()

# Create a box plot of the 'fact_length' column
sns.boxplot(data=df, y='fact_length')
plt.ylabel("Fact Length")
plt.title("Box Plot of Cat Fact Lengths")
plt.show()
```
